## Sort csv files by mean, median and std dev separately and print files along with their respective std dev, median and mean offset values plus plot them - SM - current code

```python
import os
import csv
import statistics
from statistics import mean
import matplotlib.pyplot as plt
from pathlib import Path
from statistics import mean, median, stdev
import matplotlib.pyplot as plt
from tabulate import tabulate


def sort_csv_files_by_mean(folder_path, column_index):
    files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
    sorted_files = []

    for file_name in files:
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, 'r') as file:
            csv_data = csv.reader(file)
            data_list = list(csv_data)[1:]  # Exclude the header row
            column_values = [float(row[column_index]) for row in data_list]
            column_mean = mean(column_values)
            sorted_files.append((file_name.split("/")[-1].split(".")[0].replace("-"," ").replace("_", " ").title(), column_mean))

    sorted_files.sort(key=lambda item: item[1])

    return sorted_files

def plot_sorted_data(sorted_files):
    #file_names = [item[0] for item in sorted_files]
    #file_names = [''.join([word[0] for word in item[0].split()]) for item in sorted_files]  # Extract the first letter of each word

    file_names =   [''.join([word[0] for word in item[0].split()]) for item in sorted_files]  # Replace specific phrases with their abbreviations
    mean_values = [item[1] for item in sorted_files]


    plt.figure(figsize=(30, 30))
    x_pos = range(len(file_names))
    plt.bar(x_pos, mean_values, width=0.5, align='center')
    plt.xticks(x_pos, file_names, rotation=90, fontsize=20)
    plt.xlabel('File Names', fontsize=20)
    plt.ylabel('Mean Values of offset', fontsize=20)
    plt.title('Mean Values of CSV Files', fontsize=20)
    plt.savefig("/content/drive/MyDrive/ILDARSrevised/results/ImageResults/" + 'mean_separated.svg')
    plt.tight_layout()
    plt.subplots_adjust(bottom=0.3)
    plt.show()

def sort_csv_files_by_median(folder_path, column_index):
    files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
    sorted_files2 = []

    for file_name in files:
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, 'r') as file:
            csv_data = csv.reader(file)
            data_list = list(csv_data)[1:]  # Exclude the header row
            column_values = [float(row[column_index]) for row in data_list]
            column_median = statistics.median(column_values)
            sorted_files2.append((file_name.split("/")[-1].split(".")[0].replace("-"," ").replace("_", " ").title(), column_median))

    sorted_files2.sort(key=lambda item: item[1])

    return sorted_files2

def plot_sorted_data_median(sorted_files2):
    #file_names = [item[0] for item in sorted_files2]
    file_names = [''.join([word[0] for word in item[0].split()]) for item in sorted_files2]  # Extract the first letter of each word
    median_values = [item[1] for item in sorted_files2]



    plt.figure(figsize=(30, 30))
    x_pos = range(len(file_names))
    plt.bar(x_pos, median_values, width=0.5, align='center')
    plt.xticks(x_pos, file_names, rotation=90, fontsize=20)
    plt.xlabel('File Names', fontsize=20)
    plt.ylabel('Median Values of offset', fontsize=20)
    plt.title('Median Values of CSV Files', fontsize=20)
    plt.savefig("/content/drive/MyDrive/ILDARSrevised/results/ImageResults/" + 'median_separated.svg')
    plt.tight_layout()
    plt.subplots_adjust(bottom=0.3)
    plt.show()


def sort_csv_files_by_std(folder_path, column_index):
    files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
    sorted_files3 = []

    for file_name in files:
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, 'r') as file:
            csv_data = csv.reader(file)
            data_list = list(csv_data)[1:]  # Exclude the header row
            column_values = [float(row[column_index]) for row in data_list]
            column_std = statistics.stdev(column_values)
            sorted_files3.append((file_name.split("/")[-1].split(".")[0].replace("-"," ").replace("_", " ").title(), column_std))

    sorted_files3.sort(key=lambda item: item[1])

    return sorted_files3

def plot_sorted_data_std(sorted_files3):
    #file_names = [item[0][0] for item in sorted_files3]
    file_names = [''.join([word[0] for word in item[0].split()]) for item in sorted_files3]  # Extract the first letter of each word
    std_values = [item[1] for item in sorted_files3]

    plt.figure(figsize=(30, 30))
    x_pos = range(len(file_names))
    plt.bar(x_pos, std_values, width=0.5, align='center')
    plt.xticks(x_pos, file_names, rotation=90, fontsize=20)
    plt.xlabel('File Names', fontsize=20)
    plt.ylabel('Std dev Values of offset', fontsize=20)
    plt.title('std dev of CSV Files', fontsize=20)
    plt.savefig("/content/drive/MyDrive/ILDARSrevised/results/ImageResults/" + 'stddev_separated.svg')
    plt.tight_layout()
    plt.subplots_adjust(bottom=0.3)
    plt.show()

folder_path = '/content/drive/MyDrive/ILDARSrevised/results/2023-11-03_13-49-43/'  # Replace with the actual folder path
column_index = 3  # Replace with the index of the column to calculate the mean (0-based index)

sorted_files = sort_csv_files_by_mean(folder_path, column_index)
plot_sorted_data(sorted_files)

sorted_files2 = sort_csv_files_by_median(folder_path, column_index)
plot_sorted_data_median(sorted_files2)


sorted_files3 = sort_csv_files_by_std(folder_path, column_index)
plot_sorted_data_std(sorted_files3)



def create_table(sorted_files,sorted_files2,sorted_files3):
    file_names = [item[0] for item in sorted_files]
    mean_values = [item[1] for item in sorted_files]
    median_values = [item[1] for item in sorted_files2]
    stdev_values = [item[1] for item in sorted_files3]

    table_data = list(zip(file_names, mean_values, median_values, stdev_values))
    headers = ['File Name', 'Mean', 'Median', 'Standard Deviation']
    table = tabulate(table_data, headers=headers, tablefmt='grid')



    print(table)


create_table(sorted_files,sorted_files2,sorted_files3)
```

## Helper Functions
Run this cell, before running "Bar Charts" and "Large Table"

```python
def shorten_names(algorithms_str):
  return algorithms_str \
    .replace("INVERSION","I\n\n") \
    .replace("GNOMONIC_PROJECTION","G\n\n") \
    .replace("ALL_PAIRS","A\n\n") \
    .replace("DISJOINT_PAIRS","D\n\n") \
    .replace("OVERLAPPING_PAIRS","O\n\n") \
    .replace("LARGEST_REFLECTION_CLUSTER","L\n\n") \
    .replace("NARROWEST_CLUSTER","N\n\n") \
    .replace("UNWEIGHTED_AVERAGE","U\n\n") \
    .replace("CLOSEST_LINES_EXTENDED","E\n\n") \
    .replace("CLOSEST_LINES","C\n\n") \
    .replace("MAP_TO_NORMAL_VECTOR","M\n\n") \
    .replace("REFLECTION_GEOMETRY","R\n\n") \
    .replace("WALL_DIRECTION","W\n\n") \
    .replace("-","")


def sort_csv_files(folder_path, column_index, key):
    files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
    sorted_files = []

    for file_name in files:
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, 'r') as file:
            csv_data = csv.reader(file)
            data_list = list(csv_data)[1:]  # Exclude the header row
            column_values = [float(row[column_index]) for row in data_list]
            column_sort_value = key(column_values)
            sorted_files.append((file_name.split("/")[-1].split(".")[0], column_sort_value))

    sorted_files.sort(key=lambda item: item[1])

    return sorted_files
```
